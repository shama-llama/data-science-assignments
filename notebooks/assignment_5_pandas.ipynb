### Question 3

There are columns called 'gross' and 'budget' representing the gross income and the budget for each movie. Create a new column called 'profit' which contains the gross profit (i.e. gross - budget) for each movie.

```python
mask = df['gross'].notna() & df['budget'].notna() & (df['budget'] != 0)
df.loc[mask, 'profit'] = df.loc[mask, 'gross'] - df.loc[mask, 'budget']
df.head()
```

### Question 4

Find the movies with the 10 highest budgets. Print the Name of the movie, along with the budget, gross and profit to the screen for each of these movies.

```python
top10_budgets = df.sort_values(by='budget', ascending=False).head(10)
df_to_print = top10_budgets[['movie_title', 'budget', 'gross', 'profit']].copy()
df_to_print['movie_title'] = df_to_print['movie_title'].str.ljust(30)
print("\nMovies with the 10 highest budgets:\n")
print(df_to_print.rename(columns={
    'movie_title': 'Movie Title'.ljust(30),
    'budget': 'Budget',
    'gross': 'Gross',
    'profit': 'Profit'
}))
```

> The above result shows there's a large disparity between budget and gross indicating some of the currency hasn't been normalized.

### Question 6

Check out the the groupBy and agg functions.

`pandas.DataFrame.groupby`

`pandas.core.groupby.DataFrameGroupBy.agg`

Use these to get the average budget for each director, then display the directors with the 5 largest budgets.

```python
# Compute the average budget for each director
director_avg_budget = df.groupby('director_name')['budget'].mean().reset_index()
top5_directors = director_avg_budget.sort_values(by='budget', ascending=False).head(5)
print("\nDirectors with the 5 largest average budgets:\n")
df_to_print = top5_directors[['director_name', 'budget']].copy()
df_to_print['director_name'] = df_to_print['director_name'].str.ljust(40)
print(df_to_print.rename(columns={'director_name': 'Director'.ljust(40), 'budget': 'Budget'}))
```

> In a similar fashion as the previous cell, countries from East Asia being over-represented indicates the currency hasn't been normalized.

### Distribution graphs

```python
import matplotlib.pyplot as plt
import seaborn as sns

# List of numeric columns to visualize
num_columns = ['budget', 'gross', 'profit', 'imdb_score', 'duration', 'title_year']

# Set up a grid layout: using 3 columns per row for 6 plots
n_cols = 3
n_rows = (len(num_columns) + n_cols - 1) // n_cols

fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(18, 6 * n_rows))
axes = axes.flatten()

for ax, col in zip(axes, num_columns):
    # Drop missing values
    data = df[col].dropna()
    # Convert monetary columns to millions for clarity
    if col in ['budget', 'gross', 'profit']:
        data = data / 1e6
        title = f"Distribution of {col.capitalize()} (M)"
        xlabel = f"{col.capitalize()} (M)"
    else:
        title = f"Distribution of {col.capitalize()}"
        xlabel = col.capitalize()
    
    sns.histplot(data, bins=20, kde=True, ax=ax, color='skyblue', edgecolor='black')
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel("Frequency")

# Remove any unused subplots if present
for j in range(len(num_columns), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

#### Normalize budget to USD using the conversion rate based on the 'country' column.

```python
df['budget_usd'] = df.apply(
    lambda row: (
        row['budget'] / conversion_rates.get(row['country'], 1)
        if pd.notna(row['budget']) and row['budget'] != 0 and conversion_rates.get(row['country'], 1) is not None
        else None
    ),
    axis=1
)
```

#### Recalculate profit using the normalized 'budget_usd'

```python
mask = df['gross'].notna() & df['budget_usd'].notna() & (df['budget_usd'] != 0)
df.loc[mask, 'profit'] = df.loc[mask, 'gross'] - df.loc[mask, 'budget_usd']

# Look at the updated dataframe
df.head()
```

#### Corrected list of movies with highest budget

```python
top10_budgets = df.sort_values(by='budget_usd', ascending=False).head(10)
df_to_print = top10_budgets[['movie_title', 'budget_usd']].copy()
df_to_print['movie_title'] = df_to_print['movie_title'].str.ljust(42)
print("\nMovies with the 10 highest budgets:\n")
print(df_to_print.rename(columns={
    'movie_title': 'Movie Title'.ljust(42),
    'budget_usd': 'Budget'
}))
```
